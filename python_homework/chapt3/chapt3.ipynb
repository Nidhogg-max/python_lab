### 作业三

```python
vec = [[1 , 2 , 3] , [4 , 5 , 6] , [7 , 8 , 9]]
result = []
for elem in vec:
    for num in elem:
        result.append(num)
result
```

```python
from itertools import chain
list(chain(*vec))
```

### 作业五

```python
scores = {'zhang san' : 45 , 'li si' : 78 , 'wang wu' : 40 , 'zhou liu' : 96 ,
          'zhao qi' : 65 , 'sun ba' : 90 , 'zheng jiu' : 78 , 'wu shi' : 99 , 
          'dong shiyi' : 60}
higest = max(scores.values()) 
lowest = min(scores.values())
average = sum(scores.values()) / len(scores)
higest , lowest , average
```

```python
higestPerson = [name for name , score in scores.items() if score == higest]
higestPerson
```
