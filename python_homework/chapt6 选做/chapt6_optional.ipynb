### 选做1(选必做)

```python
shu = []
for m in range(100 , 1000):
    m1 = m // 100
    m2 = (m // 10) % 10
    m3 = m % 10
    if m == m1**3 + m2**3 + m3**3:
        shu.append(m)

print(shu)
```

### 选做2

```python
#添加global后
def func(para):
    global n
    n = 3

n = 5
func()
print(n)
```

```python
#不添加global
def func(para):
    para = 3

n = 5
func(n)
print(n)
```

函数内部对变量的赋值会创建一个局部变量，原代码中的para只是局部变量，外部的n不受影响，添加global后函数内对n的赋值变成对全局变量n的修改，所以外部打印值改变
