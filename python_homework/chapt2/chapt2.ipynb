### 作业1

```python
#前后参数等长
def add(x , y):
    return x + y
list(map(add , range(5) , range(5,10)))
list(map(lambda x , y : x + y , range(5) , range(5,10)))
```

```python
#前比后长
def add(x , y):
    return x + y
list(map(add , range(8) , range(5,10)))
list(map(lambda x , y : x + y , range(8) , range(5,10)))
```

```python
#后比前长
def add(x , y):
    return x + y
list(map(add , range(5) , range(5,12)))
list(map(lambda x , y : x + y , range(5) , range(5,12)))
```

结论：map函数总是以最短的长度为标准

### 作业2

```python
seq = ['foo' , 'x41' , '?if4' , '***']
def func(x):
    return x.isalnum()
filter(func , seq)
```

```python
list(filter(func , seq))
```

```python
seq
```

```python
[x for x in seq if x.isalnum()]
```

```python
list(filter(lambda x : x.isalnum() , seq))
```

```python
list(filter(str.isalnum , seq))
```

```python
list(filter(None , [1 , 2 ,3 , 0 , 0 , 4 , 0 , 5]))
```

结论： 被isalnum()返回False过滤
