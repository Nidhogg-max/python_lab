#### 法1

```python
nums = sorted(map(int, input('Input three numbers: ').split()))
print(*nums)
```

在这里我们随机输入三个数并按照升序的输出方式输出

```python
print('Min:', min(nums), 'Max:', max(nums))
```

或者使用min(nums)和max(nums)输出最大值和最小值

#### 法2

```python
x = input('Input three numbers: ')
a, b, c = map(int, x.split())
if a > b:
    a, b = b, a
if a > c:
    a, c = c, a
if b > c:
    b, c = c, b
print(a, b, c)
```

### 选做2

```python
def func(score):
    if score > 100 or score < 0:
        return 'wrong score.must between 0 and 100.'
    elif score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'E'

def main():
    try:
        s = float(input("输入分数："))
        print("等级：", func(s))
    except ValueError:
        print("输入有效的数字")
main()
```

假设我们输入的是60，来验证一下main()是否能正常调用      可以看到调用正常且输出正确
