### 作业3

字节串和字符串的区别
字符串主要用于储存文本数据，字节串用来储存二进制数据（0-255的整数）

```python
#举例
text = "python编程"
print(type(text))
byte_data = text.encode('utf-8')
print(byte_data)
print(type(byte_data))
decoded_text = byte_data.decode('utf-8')
print(decoded_text)
print(type(decoded_text))
```
