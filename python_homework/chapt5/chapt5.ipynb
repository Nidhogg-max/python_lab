### 作业1

```python
def print_multiplication_table():
    for i in range(1, 10):
        for j in range(1, i + 1):
            print(f"{i}*{j}={i * j}", end=" ")
        print()

def main():
    print_multiplication_table()

main()
```

### 作业2

```python
for n in range(100 , 1 , -1):
    if n % 2 == 0:
        continue
    for i in range(3 , int(n**0.5) + 1 , 2):
        if n % i == 0:
            break
    else:
        print(n)
        break
```

```python
#改写打印所有素数
for n in range(2, 101):
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(n)
```

### 作业3

```python
def f():
    a , b = 1 , 1
    while True:
        yield a
        a , b = b , a + b
a = f()
for _ in range(10):
    print(a.__next__() , end=' ')
```
