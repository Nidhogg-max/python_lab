### 作业1

```python
text = "hello world, hello python"
```

```python
#find()
pos_find = text.find("hello")
print(f"find('hello') 返回第一次出现的位置: {pos_find}")
```

```python
#rfind()
pos_rfind = text.rfind("hello")
print(f"rfind('hello') 返回最后一次出现的位置: {pos_rfind}")
```

```python
#index()
pos_index = text.index("world")
print(f"index('world') 返回第一次出现的位置: {pos_index}")
try:
	pos_index = text.index("C++")
	print(f"index('C++') 返回第一次出现的位置: {pos_index}")
except ValueError:
	print("index('C++') 抛出异常: substring not found")
```

可以看到，在找不到C++这个字符串时抛出异常

```python
#count()
cnt = text.count("hello")
print(f"count('hello') 统计出现的次数: {cnt}")
```

```python
given_string = "hello world,hello python"  
result = ''.join(ch.upper() if i % 2 == 0 else ch for i, ch in enumerate(given_string))
print("原字符串:", given_string)
print("转换后:", result)
```

### 作业3

```python
#原代码
import string
def kaisa(s, k):
    lower = string.ascii_lowercase   
    upper = string.ascii_uppercase    
    before = string.ascii_letters    
    after = lower[k:] + lower[:k] + upper[k:] + upper[:k]
    table = ' '.maketrans(before, after)    
    return s.translate(table)

s = "Python is a great programming language. I like it!"
print(kaisa(s, 3))
```

```python
#修改k值（可自定义k）
import string

def kaisa(s, k):
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase
    before = string.ascii_letters
    k = k % 26
    after = lower[k:] + lower[:k] + upper[k:] + upper[:k]
    table = str.maketrans(before, after)   
    return s.translate(table)

s = input("请输入要加密的字符串: ")
k = int(input("请输入凯撒加密的偏移量 K: "))
print("加密结果:", kaisa(s, k))
```

当前为输入Hollow world，k = 5 时的加密结果
