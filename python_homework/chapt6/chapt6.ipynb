### 作业1

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

def main():
    try:
        num = int(input("输入一个整数"))
        if is_prime(num):
            print(f"{num} 是一个素数")
        else:
            print(f"{num} 不是一个素数")
    except ValueError:
        print("请输入一个有效的整数")
        

main()
```

### 作业2

```python
def stats(*numbers):
    if not numbers:
        print("没有输入数字")
        return
    
    max_num = max(numbers)
    total = sum(numbers)
    print(f"最大值: {max_num}")
    print(f"总和: {total}")

if __name__ == "__main__":
    stats(3 , 7 , 2 , 9 , 5)
    stats(10 , -2 , 0 , 8 , 4)
    stats()
```
