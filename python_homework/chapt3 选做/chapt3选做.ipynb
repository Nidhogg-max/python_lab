### 选做1

列表推导式法

```python
import random
x = [10 , 2 , 5 , 16 , 8 , 19 , 18 , 11 , 9 , 3 , 1 , 0 , 14 , 6 , 17]
x1 = [i for i , val in enumerate(x) if val > 12]
print("大于12" , x1)
```

```python
x2 = [i for i , val in enumerate(x) if 5 < val < 12]
print("5到12之间" , x2)
```

Numpy数组法

```python
import numpy as np
x = [10 , 2 , 5 , 16 , 8 , 19 , 18 , 11 , 9 , 3 , 1 , 0 , 14 , 6 , 17]
a = np.array(x)
x1 = a > 12
x1 = np.where(x1)[0].tolist()
print("大于12" ,  x1)
```

```python
x2 = (a > 5)&(a < 12)
x2 = np.where(x2)[0].tolist()
print("5到12之间" , x2)
```

### 选做2

map函数法

```python
vec = [[1 ,2 , 3] , [4 ,5 ,6] , [7 ,8 ,9]]
result = list(map(lambda x : x[0] , vec))
print(result)
```

numpy数组

```python
import numpy as np
vec = [[1 ,2 , 3] , [4 ,5 ,6] , [7 ,8 ,9]]
img = np.array(vec , dtype=np.uint8)
a = img[: , 0]
result = a.tolist()
print(result)
```
