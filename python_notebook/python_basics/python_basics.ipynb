##### 2.3.1 消除空白 lstrip()和rstrip()的用法

```python
language = " python "
print(language,1)
print(language.rstrip(),1)
print(language,1)
```

本程序可以看出这个字符串右端有一个空白，我们在变量名后添加.rstrip()，空白格被消除，但再次打印字符串，空白格依旧存在，这是因为这种改变通常是一次性的

```python
language = "python "
print(language,1)
language = language.rstrip()
print(language,1)
print(language,1)
```

正确的做法应该为将添加.rstrip()的变量重新赋给原变量，这样能够永久消除字符串右端的空白 【注意rstrip()只能消除字符串右端空格】由于功能类似，lstrip()不作演示了

```python
language = " python "
print(language,1)
language = language.strip()
print(language,1)
```

该代码给出了变量加.strip()的方式，可以简单理解为消除字符串两端的所有空白格，是rstrip()和lstrip()功能的叠加

##### 2.3.2 删除前缀  removeprefix()用法

```python
nostrach_url = 'https://nostrash.com'
print(nostrach_url)
nostrach_url.removeprefix('https:;//')
print(nostrach_url)
print(nostrach_url.removeprefix('https://'))
nostrach_url = nostrach_url.removeprefix('https://')
print(nostrach_url)
```

本程序可以看到，.removeprefix()和上一小节的rstrip()与lstrip()非常相似，这种改变也是一次性的，需要对原变量重新赋值才能永久改变，另外需要注意，这个语法的括号里需要用单引号引出需要删除的前缀内容

#### 4.3 管理列表

##### 4.3.1 列表永久排序  【.sort()的用法】

```python
bicycles = ['bike' , 'cannondale' , 'add' , 'dake']
print(bicycles)
bicycles.sort()                 #直接对列表名使用.sort()来进行排序
print(bicycles)
bicycles.sort(reverse = True)   #括号里的内容可以使列表按照逆元素首字母顺序排列
print(bicycles)
#永久性检验
print(bicycles)
```

本程序分别展示了使用.sort()和.sort(reverse = True)对列表内元素进行按照首字母顺序和按照首字母逆序排列，并且这种排列是永久的

##### 4.3.2 列表临时排序 【.sorted()的用法】

```python
bicycles = ['bike' , 'cannondale' , 'add' , 'dake']
print(bicycles)
print(sorted(bicycles))
print(bicycles)
print(sorted(bicycles,reverse = True))
```

本程序展示sorted()的用法，可以看出，sorted()只能短暂改变列表顺序，原列表不受影响，逆序排列同sort()一样添加reverse = True即可

##### 4.3.3 反向打印列表  【.reverse()的用法】

```python
bicycles = ['bike' , 'cannondale' , 'add' , 'dake']
bicycles.reverse()
print(bicycles)
```

本程序显示reverse命令只能将原本的列表元素反向打印出来，不具有按照首字母排序的功能，并且这种改变是永久性的

### **5.操作列表（for循环）**

#### 5.1 遍历列表

```python
cassell = ['lumingfei' , 'chuzihang' , 'kaisa' , 'nuonuo']
for student in cassell:
    print(student)
```

这就是一个简单的遍历列表的for循环语句，注意for语句后的语句在循环内需要缩进

```python
cassell = ['lumingfei' , 'chuzihang' , 'kaisa' , 'nuonuo']
for student in cassell:
    print(f"{student.title()} is a good student")  #使用.title()进行首字母大写
    print(f"It is a good student who is {student}\n")  #\n在此处用来添加一个空行
print("Congratulation!")
```

```python
for student in range(1,5):
    print(student)
for cassell in range(6):
    print(cassell)
```

#### 6.4使用if处理列表

```python
requested_toppings = ['mushrooms' , 'green peppers' , 'extra cheese']
for requested_topping in requested_toppings:
    if requested_topping == 'green peppers':
        print("Sorry,we are out of green peppers right now")
    else:
        print(f"Adding {requested_topping}")
print("\nFinshed making your pizza")
```

如本程序所示，可以在for循环里添加if语句对列表中的特定元素进行判断，从而执行不同于其他元素的指令

```python
requested_toppings = []
if requested_toppings:
    for requested_topping in requested_toppings:
        print(f"Adding {requested_topping}")
    print("\nFinshed making your pizza")
else:
    print("Are you sure you want a plain pizza?")
```

这里对本代码进行一个解释，将列表名用作条件表达式时，当列表中元素为空返回False，进入到else中执行命令，如果至少包含一个元素返回值，判断结果为True，则会和上一个程序一样返回语句，**另外值得一提的是，对于各种空值的判断，都会返回False**

#### 6.5 使用多个列表

```python
available_toppings = ['mushrooms' , 'olives' , 'green peppers' , 'pepperoni' , 'pineapple' , 'extra cheese']
requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for requested_topping in requested_toppings:
    if requested_topping in available_toppings:
        print(f"Adding {requested_topping}")
    else:
        print(f"Sorry,we don't have {requested_topping}")

print("\nFinshed making your pizza")
```

定义两个列表，一个原始列表，一个预期列表，对预期列表进行for循环逐一列举，并通过if判断语句和原始列表进行对比，若能够找到则执行if语句后的命令，若未在原始列表里找到，则执行else里的命令

#### 8.2 循环结构

```python
current_number = 1
while current_number <= 5:
    print(current_number)
    current_number += 1
```

首先将1赋给变量，即变量从1开始数，下面进入while循环，循环条件为变量值小于等于5,并且在不满足条件之前，每循环一次打印一次变量值，打印后变量值加1,直到值大于5停止循环

```python
prompt = "\nTell me something,and i will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program."

message  = ""
while message != 'quit':
    message = input(prompt)  #显示提示prompt，并且让用户输入一个字符串
    if message != 'quit':    #这里的判断语句防止最后将quit也打印出来
        print(message)
```

```python
#使用标志来进行while循环
prompt = "\nTell me something,and i will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program."

active = True
message  = ""
while active:
    message = input(prompt)  #显示提示prompt，并且让用户输入一个字符串
    if message == 'quit':
        active = False
    else:  
        print(message)
```

在程序面临多个可能终止的条件时，可以选择设置一个开关，当满足终止的条件出现时将开关关上，其他时候开关保持打开

##### 8.2.1 break语句推出循环

```python
#使用break进行退出循环操作
prompt = "\nPlease enter the name of the city you have visited:"
prompt += "\n(Enter 'quit' when you are finished.)"

while True:
    city = input(prompt)

    if city == 'quit':
        break
    else:
        print(f"I'd love to go to {city.title()}!")
```

当while True打开循环不断运行，直到输入值为quit时，进行break语句退出循环

##### 8.2.2 continue语句

```python
current_number = 0
while current_number < 10:
    current_number += 1
    if current_number %2 == 0:
        continue
    print(current_number)
```

可以看到continue语句和break语句之间的差别，continue在满足if条件判断语句时继续执行while循环，在不满足时跳过本轮循环，但不退出while循环语句，继续进行下一次循环，而break语句在不满足继续循环的条件时直接退出整个循环，不再进行下一轮

```python
unconfirmed_users = ['alice' , 'brian' , 'candace']
confirmed_users = []

while unconfirmed_users:
    current_user = unconfirmed_users.pop()

    print(f"Verifying user: {current_user.title()}")
    confirmed_users.append(current_user)
    print ("\nThe following users have been confirm:")
    for confirmed_user in confirmed_users:
        print(confirmed_user.title())
```

##### 9.1.1实参和形参

```python
#向函数传递参数
def great_user(username):
    print(f"Hello,{username.title()}")

great_user('cassell')
```

可以看到，本程序中的username为形式参数（只是定义函数时为了保证程序完整性设定的虚拟信息）cassell为实际参数，即使用这个函数时实际输入的值，在一定程度上，实际参数为替换定义函数体内形式参数的值

```python
#位置实参
def describe_pet(animal_type , pet_name):
    print(f"\nI have a {animal_type}")
    print(f"{animal_type.title()}'s name is {pet_name}")

describe_pet('cat' , 'maomao')
describe_pet('dog' , 'xiaogou')
```

本程序相比上一个程序没有本质上的逻辑差异，只是在当前程序中多了一个参数(即形式参数)，一般来说，最常见的函数都具有两个形式参数。

```python
#关键字实参
def describe_pet(animal_type , pet_name):
    print(f"\nI have a {animal_type}")
    print(f"{animal_type.title()}'s name is {pet_name}")

describe_pet(animal_type = 'cat' , pet_name = 'maomao')
```

函数定义部分与上述无异，但在调用函数时直接将形参和实参联系起来，这样做的好处十分明显，可以不用刻意注意参数顺序

```python
#为形参提供默认值
def describe_pet(animal_type , pet_name = 'maomao'):
    print(f"\nI have a {animal_type}")
    print(f"{animal_type.title()}'s name is {pet_name}")

describe_pet(animal_type = 'cat')
```

在定义函数时直接对参数进行定义，在调用函数时如果没有对该形参进行赋值，则会显示函数定义时的预设值，**并且需要先列出没有默认值的**

##### 9.2.1 传递任意数量的实参

```python
def make_pizza(*toppings):
    print("\nMaking a pizza with the following toppings:")
    for topping in toppings:
        print(f"- {topping}")

make_pizza('pepperoni')
make_pizza('mushrooms' , 'green peppers' , 'extra cheese')
```

本程序可以看到，没有像之前的代码一样，一个实参对应一个形参，而是多个实参对应一个形参，只要在形参前加一个 **‘*’** 即可

##### 9.2.2 结合使用位置实参和任意数量实参

```python
def make_pizza(size , *toppings):
    print(f"\nMaking a {size}-inch pizza with the following toppings:")
    for topping in toppings:
        print(f"- {topping}")

make_pizza(16 , 'pepperoni')
make_pizza(12 , 'mushrooms' , 'green peppers' , 'extra cheese')
```

函数将第一个值给size，其他所有值赋给*toppings
